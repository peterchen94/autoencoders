```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, auc
from sklearn.ensemble import IsolationForest, RandomForestClassifier
from sklearn.metrics import average_precision_score,roc_curve, auc, f1_score, precision_recall_curve, precision_score,recall_score,accuracy_score,confusion_matrix
import sys
from inspect import signature
from sklearn.linear_model import LogisticRegression
import xgboost as xgb
```

```python
def plot_roc(y_true,y_pred):
    fpr, tpr, thresholds = roc_curve(y_true,y_pred)
    roc_auc = auc(fpr,tpr)

    plt.figure()
    lw = 2
    plt.plot(fpr, tpr, color='darkorange',
             lw=lw, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC for reconstruction error classification')
    plt.legend(loc="lower right")
    plt.show()

def plot_pr_curve(y_true,y_pred):
    precision, recall, _ = precision_recall_curve(y_true, y_pred)
    average_precision = average_precision_score(y_true, y_pred)

    plt.figure()
    area = auc(recall, precision)

    step_kwargs = ({'step': 'post'}
                   if 'step' in signature(plt.fill_between).parameters
                   else {})
    plt.step(recall, precision, color='b', alpha=0.2,
             where='post')
    plt.fill_between(recall, precision, alpha=0.2, color='b', **step_kwargs)

    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.ylim([0.0, 1.05])
    plt.xlim([0.0, 1.0])

    plt.title('Precision-Recall curve: AUC={0:0.2f}'.format(
          area))

    plt.show()
```

## Read dataset

```python
DATA_DIR = '/Users/chenpete/Documents/code/unsupervised_anomaly_detection/data/'

df = pd.read_csv(DATA_DIR + 'papermill_coding_challenge.csv')
```

```python
param_cols = ['x1', 'x2', 'x3', 'x4', 'x5', 'x6', 'x7',
       'x8', 'x9', 'x10', 'x11', 'x12', 'x13', 'x14', 'x15', 'x16', 'x17',
       'x18', 'x19', 'x20']

y_col = 'y'
```

### Visualize class imbalance

```python
sns.countplot(df['y'])
```

# visualize missing data

```python
sns.heatmap(df.isnull(),cbar ='false')
```

### Fill in missing data with foward fill - time series

```python
df.fillna(method = 'ffill',inplace=True)
```

### Univariate time series plotting - check for outliers

```python
for col in param_cols:
    plt.figure()
    plt.title(col)
    plt.plot(df[col])
```

### Univariate plotting - see if failures and non-failures are clearly separated

```python
for col in param_cols:
    plt.figure()
    plt.title(col)
    sns.distplot(df[df['y']==0][col],color='blue',kde =False)
    sns.distplot(df[df['y']==1][col],color='red',kde =False)
```

### remove outliers with 3 sigma quantile, forward fill data afterwards

for col in param_cols:
    q_upper = df[col].quantile(0.995)
    q_lower = df[col].quantile(0.005)
    df[col] = np.where(df[col] > q_upper,np.nan,df[col])
    df[col] = np.where(df[col] < q_lower,np.nan,df[col])

df.fillna(method='ffill',inplace=True)

### X y split

```python
X = df[param_cols]
y= df[y_col]
```

### Train/test split

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = .4,stratify = y)
```

```python
y_test.value_counts()
```

```python
y_train.value_counts()
```

### Standard scaler

```python
sc = StandardScaler()
sc.fit(X_train)
X_train_s = sc.transform(X_train)
X_test_s = sc.transform(X_test)
```

### Logistic Regression 

```python
clf_lr = LogisticRegression()
clf_lr.fit(X_train_s,y_train)

y_pred_lr_proba = clf_lr.predict_proba(X_test_s)[:,1]
y_pred_lr = clf_lr.predict(X_test_s)

plot_roc(y_test,y_pred_lr_proba)
plot_pr_curve(y_test,y_pred_lr_proba)

plt.figure()
plt.title('separation of failures and non-failures')
sns.distplot(y_pred_lr_proba[y_test ==0],kde=False,color='blue')
sns.distplot(y_pred_lr_proba[y_test ==1],kde=False,color='red')
plt.show()

```

```python
y_threshold = np.where(y_pred_lr_proba >.13 ,1,0)

print('f1 score is {:.2f}'.format(f1_score(y_test,y_threshold)))
print('precision score is {:.2f}'.format(precision_score(y_test,y_threshold)))
print('recall score is {:.2f}\n'.format(recall_score(y_test,y_threshold)))
print('confusion_matrix is \n{}'.format(confusion_matrix(y_test,y_threshold)))
```
